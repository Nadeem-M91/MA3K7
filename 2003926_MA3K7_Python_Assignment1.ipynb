```python
import numpy as np
```

For the first part of my code I will run each starting pair to see the lengths of the bracelet chain that will be created by running our formula.


```python
# Define the chain and chain length
def calculate_chain_length(starting_pair):
    current_pair = starting_pair
    chain = [current_pair]

    # Create our formula to work out the next number from the previous two numbers    
    while True:
        next_number = np.sum(current_pair) % 10
        new_pair = (current_pair[1], next_number)

        # Stop chain if we reach our starting pair
        if new_pair == starting_pair:
            break
        
        chain.append(new_pair)
        current_pair = new_pair
    
    return len(chain)

# Now do the same for every pair in the range of (0,0) to (9,9)
def main():
    for i in range(10):
        for j in range(10):
            starting_pair = (i, j)
            chain_length = calculate_chain_length(starting_pair)
            print(f"For Starting pair {starting_pair}: the Chain length is = {chain_length}")

if __name__ == "__main__":
    main()
```

For this second section I will find how many chains are of each length. This is to create a more consise reading for the large number of data above which may not be so easy to read.

```python
# Similar to first part find the lengths of each chain.
def generate_chain_length(starting_pair):
    current_pair = starting_pair
    chain = [current_pair]
    
    while True:
        next_number = (current_pair[0] + current_pair[1]) % 10
        current_pair = (current_pair[1], next_number)
        
        if current_pair == starting_pair:
            break
            
        chain.append(current_pair)
    
    return len(chain)


# For each chain length calculate how many chains there are
def main1():
    chain_lengths_count = {}
    
    for i in range(100):
        pair = (i // 10, i % 10)
        chain_length = generate_chain_length(pair)
        
        if chain_length not in chain_lengths_count:
            chain_lengths_count[chain_length] = 1
        else:
            chain_lengths_count[chain_length] += 1
    
    for length, count in sorted(chain_lengths_count.items()):
        print(f" For Chain length {length}: there are {count} chains")

                
if __name__ == "__main__":
    main1()
```
